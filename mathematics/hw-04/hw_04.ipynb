## Task 1: Log Error Analytics

Given: $|\Omega| = 500$, DB only = 85, NET only = 60, both DB and NET = 35.

Total DB errors: $|A| = 85 + 35 = 120$, total NET errors: $|B| = 60 + 35 = 95$.

**P(A) — probability of a DB error:**
$$P(A) = \frac{|A|}{|\Omega|} = \frac{120}{500} = 0.24$$

**P(A ∪ B) — probability of DB or NET:**
$$P(A \cup B) = P(A) + P(B) - P(A \cap B) = \frac{120}{500} + \frac{95}{500} - \frac{35}{500} = \frac{180}{500} = 0.36$$

**P(A \ B) — probability of DB but not NET:**
$$P(A \setminus B) = \frac{85}{500} = 0.17$$

```python
import math
from itertools import combinations

total = 500
db_only, net_only, both = 85, 60, 35

db_total = db_only + both
net_total = net_only + both

p_a = db_total / total
p_b = net_total / total
p_ab = both / total

p_a_or_b = p_a + p_b - p_ab
p_a_not_b = db_only / total

print(f"P(A) — DB error:        {p_a:.4f}")
print(f"P(A ∪ B) — DB or NET:   {p_a_or_b:.4f}")
print(f"P(A \\ B) — DB not NET:  {p_a_not_b:.4f}")
```

$$P(\text{exactly 2 good}) = \frac{672}{1320} = \frac{28}{55} \approx 0.5091$$

**Combinatorial verification:**
$$P = \frac{C_8^2 \cdot C_4^1}{C_{12}^3} = \frac{28 \cdot 4}{220} = \frac{112}{220} = \frac{28}{55} \approx 0.5091$$

```python
n_good, n_defect, n_total = 8, 4, 12

# all 3 good — multiplication rule
p_all_good_mult = (8/12) * (7/11) * (6/10)

# all 3 good — combinatorial
p_all_good_comb = math.comb(8, 3) / math.comb(12, 3)

# exactly 2 good — multiplication rule
p_ggd = (8/12) * (7/11) * (4/10)
p_gdg = (8/12) * (4/11) * (7/10)
p_dgg = (4/12) * (8/11) * (7/10)
p_two_good_mult = p_ggd + p_gdg + p_dgg

# exactly 2 good — combinatorial
p_two_good_comb = (math.comb(8, 2) * math.comb(4, 1)) / math.comb(12, 3)

print(f"All 3 good (multiplication rule): {p_all_good_mult:.4f}")
print(f"All 3 good (combinatorial):       {p_all_good_comb:.4f}")
print(f"Exactly 2 good (multiplication):  {p_two_good_mult:.4f}")
print(f"Exactly 2 good (combinatorial):   {p_two_good_comb:.4f}")
```

## Task 3: Distributed System Reliability

Failure probabilities: $p_A=0.02,\ p_B=0.05,\ p_C=0.03,\ p_D=0.04$.
Working probabilities: $\bar{p}_A=0.98,\ \bar{p}_B=0.95,\ \bar{p}_C=0.97,\ \bar{p}_D=0.96$.

**1. All 4 nodes working** (independent events):
$$P(\text{all work}) = 0.98 \cdot 0.95 \cdot 0.97 \cdot 0.96 \approx 0.8669$$

**2. At least one fails** (complementary event):
$$P(\text{at least one fails}) = 1 - P(\text{all work}) \approx 1 - 0.8669 = 0.1331$$

**3. Exactly two fail** — all $C_4^2 = 6$ combinations:

| Pair | Probability |
|------|-------------|
| A, B | $0.02 \cdot 0.05 \cdot 0.97 \cdot 0.96 \approx 0.000931$ |
| A, C | $0.02 \cdot 0.03 \cdot 0.95 \cdot 0.96 \approx 0.000547$ |
| A, D | $0.02 \cdot 0.04 \cdot 0.95 \cdot 0.97 \approx 0.000737$ |
| B, C | $0.05 \cdot 0.03 \cdot 0.98 \cdot 0.96 \approx 0.001411$ |
| B, D | $0.05 \cdot 0.04 \cdot 0.98 \cdot 0.97 \approx 0.001901$ |
| C, D | $0.03 \cdot 0.04 \cdot 0.98 \cdot 0.95 \approx 0.001117$ |

$$P(\text{exactly 2 fail}) \approx 0.006645$$

```python
failures = {'A': 0.02, 'B': 0.05, 'C': 0.03, 'D': 0.04}
working = {node: 1 - p for node, p in failures.items()}

# all 4 working
p_all_work = 1
for p in working.values():
    p_all_work *= p
print(f"All 4 nodes working:      {p_all_work:.6f}")

# at least one fails
p_at_least_one = 1 - p_all_work
print(f"At least one fails:       {p_at_least_one:.6f}")

# exactly two fail
nodes = list(failures.keys())
p_exactly_two = 0
for pair in combinations(nodes, 2):
    p = 1
    for node in nodes:
        p *= failures[node] if node in pair else working[node]
    print(f"  Pair {pair} fails: {p:.6f}")
    p_exactly_two += p

print(f"Exactly two fail:         {p_exactly_two:.6f}")
```

## Task 4: Traffic Source Analysis

Sources and conversion rates:

| Source | $P(H_i)$ | $P(A \mid H_i)$ |
|--------|----------|-----------------|
| Search ads | 0.50 | 0.04 |
| Social media | 0.30 | 0.02 |
| Email | 0.20 | 0.08 |

**1. Total probability of purchase:**
$$P(A) = \sum_i P(H_i) \cdot P(A \mid H_i) = 0.5 \cdot 0.04 + 0.3 \cdot 0.02 + 0.2 \cdot 0.08 = 0.042$$

**2. Probability buyer came from email (Bayes' theorem):**
$$P(H_3 \mid A) = \frac{P(H_3) \cdot P(A \mid H_3)}{P(A)} = \frac{0.2 \cdot 0.08}{0.042} = \frac{0.016}{0.042} \approx 0.3810$$

```python
prior_probs = [0.50, 0.30, 0.20]
conversion_rates = [0.04, 0.02, 0.08]
source_names = ["Search ads", "Social media", "Email"]

def bayes_source(prior_probs, conversion_rates):
    joint = [p * r for p, r in zip(prior_probs, conversion_rates)]
    total = sum(joint)
    return [j / total for j in joint]

p_purchase = sum(p * r for p, r in zip(prior_probs, conversion_rates))
print(f"Total probability of purchase: {p_purchase:.4f}")

posteriors = bayes_source(prior_probs, conversion_rates)
print("\nPosterior probabilities given a purchase was made:")
for name, post in zip(source_names, posteriors):
    print(f"  {name}: {post:.4f}")
```
